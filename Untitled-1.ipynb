```python
import pandas as pd

print("전체 데이터에서 기계 출석부와 랭킹을 뽑아내는 중이야! 조금만 기다려줘 ⏳")

# 1. 꼭 필요한 칸('기계 이름', '진짜 쓴 전기')만 쏙 뽑아서 초스피드로 불러오기
# 주의: '진짜_파일_이름.csv' 부분을 네 파일 이름으로 꼭 바꿔줘!
df = pd.read_csv('rtu_data_full.csv', usecols=['module(equipment)', 'activePower'])

# 2. 기계가 총 몇 종류인지 출석 부르기
machine_count = df['module(equipment)'].nunique()
print(f"\n=== 1. 📢 기계 출석부 ===")
print(f"우리 데이터 센터에는 총 {machine_count}종류의 기계가 일하고 있어!")

# 3. 전기를 제일 많이 쓴 기계 랭킹 TOP 10 뽑기 (누적 사용량 기준)
# 기계별로 전기를 얼마나 썼는지 다 더한 다음, 1등부터 10등까지 줄을 세우는 코드야!
print("\n=== 2. 🏆 전기 먹방 랭킹 TOP 10 ===")
ranking = df.groupby('module(equipment)')['activePower'].sum().sort_values(ascending=False).head(10)

# 4. 결과 보여주기
print(ranking)
print("\n🎉 랭킹 계산 완료! 결과를 꼭 복사해서 에이전트에게 알려주세요!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 1. operation 값의 종류부터 확인 (0/1인지, 다른 코드인지)
df = pd.read_csv('rtu_data_full.csv', 
                  usecols=['module(equipment)', 'localtime', 'operation', 'activePower', 
                           'powerFactorR', 'powerFactorS', 'powerFactorT'])

print("operation 값 종류:", df['operation'].unique())
print(df.groupby('operation')['activePower'].describe())
```

```python
df_crusher = df[df['module(equipment)'] == '4(분쇄기(1))'].copy()
df_crusher['localtime'] = pd.to_datetime(df_crusher['localtime'].astype(str), format='%Y%m%d%H%M%S')

# operation=0(정지 상태)인데 activePower가 남아있는 구간 찾기
idle_waste = df_crusher[(df_crusher['operation'] == 0) & (df_crusher['activePower'] > 100)]
print(f"⚠️ 정지 상태인데 전력 낭비 중인 데이터 개수: {len(idle_waste)}")
print(idle_waste[['localtime', 'activePower']].head(20))

# 역률(효율) 평균도 같이 확인
df_crusher['avg_powerFactor'] = df_crusher[['powerFactorR','powerFactorS','powerFactorT']].mean(axis=1)
print("\noperation별 평균 역률:")
print(df_crusher.groupby('operation')['avg_powerFactor'].mean())
```

```python
df_all = pd.read_csv('rtu_data_full.csv', 
                      usecols=['module(equipment)', 'localtime', 'activePower'])
df_all['localtime'] = pd.to_datetime(df_all['localtime'].astype(str), format='%Y%m%d%H%M%S')
df_all['hour'] = df_all['localtime'].dt.hour

# 설비별로 "시간대에 따른 변동폭(표준편차)"을 비교
variation_by_machine = df_all.groupby(['module(equipment)', 'hour'])['activePower'].mean().unstack()
variation_range = variation_by_machine.max(axis=1) - variation_by_machine.min(axis=1)
print("설비별 시간대 평균의 변동폭(최대-최소):")
print(variation_range.sort_values(ascending=False))
```

```python
import pandas as pd

df = pd.read_csv('rtu_data_full.csv', 
                  usecols=['module(equipment)', 'voltageR','voltageS','voltageT',
                           'currentR','currentS','currentT','activePower',
                           'powerFactorR','powerFactorS','powerFactorT'])

# 전압 불균형율 계산: (최대-최소)/평균 * 100 (%)
df['v_avg'] = df[['voltageR','voltageS','voltageT']].mean(axis=1)
df['v_max'] = df[['voltageR','voltageS','voltageT']].max(axis=1)
df['v_min'] = df[['voltageR','voltageS','voltageT']].min(axis=1)
df['voltage_imbalance'] = (df['v_max'] - df['v_min']) / df['v_avg'] * 100

# 전류 불균형율도 동일하게
df['c_avg'] = df[['currentR','currentS','currentT']].mean(axis=1)
df['c_max'] = df[['currentR','currentS','currentT']].max(axis=1)
df['c_min'] = df[['currentR','currentS','currentT']].min(axis=1)
df['current_imbalance'] = (df['c_max'] - df['c_min']) / df['c_avg'] * 100

# 설비별 평균 불균형율 랭킹 (전류가 특히 중요! 전류 불균형이 손실에 더 직결됨)
print("=== 설비별 평균 전류 불균형율 (%) ===")
print(df.groupby('module(equipment)')['current_imbalance'].mean().sort_values(ascending=False))

print("\n=== 설비별 평균 전압 불균형율 (%) ===")
print(df.groupby('module(equipment)')['voltage_imbalance'].mean().sort_values(ascending=False))

print("\n=== 설비별 평균 역률 (%) ===")
avg_pf = df[['powerFactorR','powerFactorS','powerFactorT']].mean(axis=1)
df['avg_pf'] = avg_pf
print(df.groupby('module(equipment)')['avg_pf'].mean().sort_values())
```
